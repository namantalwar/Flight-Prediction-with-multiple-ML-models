<br>
<br>
<h3>Cleaning The Datasets</h3>
<br>
Extracting the Required Colunns

```python
data = data[["Year", 'International Passenger Revenue ($)','Domestic Passenger Revenue ($)','Total Fuel Expenses ($)', 
             'Fuel Expense per Enplaned Passenger ($)', 'Fuel price per Gallon ($)','Total Fuel in Gallons']]
data.head()
```

```python
data2.columns
```

```python
data2 = data2[["year","atf"]]
data2.head()
```

```python
data2.rename(columns={"year":"Year","atf":"ATF Imported"},inplace = True)
data2.head()
```

<br>
<br>
<h3>Merging The 2 Datasets</h3>

```python
datafinal = pd.merge(data,data2,on = "Year" )
datafinal.head()
```

<br>
<h3>Calculating The Total Passenger Revenue</h3>
<br>
By Adding The Revenue from International Passengers and Domestic Passengers 

```python
datafinal['International Passenger Revenue ($)'] = datafinal['International Passenger Revenue ($)'].fillna(0)
datafinal["Domestic Passenger Revenue ($)"]=datafinal['Domestic Passenger Revenue ($)'].fillna(0)
datafinal["Total Revenue ($)"] = datafinal['International Passenger Revenue ($)']+datafinal["Domestic Passenger Revenue ($)"]
datafinal.head()
```
